```python
Create a  condition = 'LoanStatus'== 'Completed'| 'LoanStatus' == 'Defaulted' |\'LoanStatus' == 'Chargedoff'
```

```python

def loan_status(x):
    if x=="Completed":
        return 1
    if x=="Defaulted":
        return 1
    if x=="Chargedoff":
        return 1
    else:
        return 0
df2["LoanStatus_new"]=df2["LoanStatus"].apply(loan_status)
```
